## ðŸ§© Eksik DeÄŸer Analizi

```python
for col in df.columns:
    eksik = df[col].isnull().sum()
    if eksik > 0:
        print(f"{col} â†’ Eksik deÄŸer sayÄ±sÄ±: {eksik}")
```

### ðŸ”§ Eksik DeÄŸerlerin Ele AlÄ±nmasÄ± â€“ Arrival Delay in Minutes

Veri setinde yalnÄ±zca `Arrival Delay in Minutes` deÄŸiÅŸkeninde eksik deÄŸer gÃ¶zlemlenmiÅŸtir (toplam 393 satÄ±r). Bu durum, veri setinin genelinde eksik deÄŸer sorununun Ã§ok sÄ±nÄ±rlÄ± olduÄŸunu gÃ¶stermektedir.

Bu eksik deÄŸerleri ele almak iÃ§in birkaÃ§ yaklaÅŸÄ±m deÄŸerlendirilmiÅŸtir. Ã–ncelikle, sÃ¶z konusu deÄŸiÅŸkenin ortalamasÄ± **15.09 dakika** iken, medyanÄ± **0.00 dakika** olarak hesaplanmÄ±ÅŸtÄ±r. Bu durum, verinin saÄŸa Ã§arpÄ±k (positively skewed) daÄŸÄ±ldÄ±ÄŸÄ±nÄ± ve yÃ¼ksek uÃ§ deÄŸerler iÃ§erdiÄŸini gÃ¶stermektedir. Bu nedenle, ortalama ile doldurma yÃ¶ntemi veriyi bozabilir.

Bu sebeplerden dolayÄ±, eksik deÄŸerlerin **medyan ile doldurulmasÄ±** tercih edilmiÅŸtir. BÃ¶ylece verinin daÄŸÄ±lÄ±m yapÄ±sÄ± korunmuÅŸ ve analiz sonuÃ§larÄ± daha gÃ¼venilir hale gelmiÅŸtir.


```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median())
```

```python
df.isnull().sum()
```

### ðŸ”§ AykÄ±rÄ± DeÄŸer Analizi


AykÄ±rÄ± deÄŸer analizinde sadece sayÄ±sal deÄŸiÅŸkenler ele alÄ±nÄ±r.

```python
for col in numeric_cols:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    alt_sinir=q1-1.5*iqr
    ust_sinir=q3+1.5*iqr
    aykiri_sayi = df[(df[col] < alt_sinir) | (df[col] > ust_sinir)].shape[0]
    oran = (aykiri_sayi / df.shape[0]) * 100

    print(f"ðŸ“Œ {col}")
    print(f"Q1: {q1:.2f} | Q3: {q3:.2f} | IQR: {iqr:.2f}")
    print(f"Alt SÄ±nÄ±r: {alt_sinir:.2f} | Ãœst SÄ±nÄ±r: {ust_sinir:.2f}")
    print(f"AykÄ±rÄ± DeÄŸer SayÄ±sÄ±: {aykiri_sayi} ({oran:.2f}%)\n")
```
